# 2B - INSURANCE SURVEY

## Preprocessing Data

```python
df = pd.read_excel("/content/drive/MyDrive/Datasets/Data Science Research/IS403/0. Insurance Survey.xlsx")
df
```

```python
df.columns = list(df.iloc[1])
df = df[2:26]
```

```python
df.reset_index(inplace=True)
df = df.iloc[:, 1:]
```

```python
df
```

```python
df['Satisfaction* '] = df['Satisfaction* '].astype(float, errors = 'raise')
```

```python
df.info()
```

## Test

```python
li_group = [d["Satisfaction* "] for _, d in df.groupby("Education")]
```

```python
stat_levene, p_value_levene = stats.levene(*li_group, center="mean")
```

```python
print(f"stats levene = {stat_levene}")
print(f"p_value levene = {p_value_levene}")
```

```python
import scipy.stats as stats
```

```python
fvalue_anova, pvalue_anova = stats.f_oneway(*li_group)
```

```python
print(f"f-value = {fvalue_anova}")
print(f"p-value = {pvalue_anova}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
tukey = pairwise_tukeyhsd(endog=df["Satisfaction* "], groups=df["Education"], alpha=0.05)
print(tukey)
```

# 2C - ENERGY DRINK SURVEY
